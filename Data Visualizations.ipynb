# Dataset of OCD Patients
following is the link for the dataset:
https://www.kaggle.com/datasets/ohinhaque/ocd-patient-dataset-demographics-and-clinical-data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('ocd_patient_dataset.csv')
print(data)
```

### Dropping extra columns

```python
data.drop(['Patient ID','Ethnicity'],axis=1,inplace=True)
```

```python
df = data.head(100)
df
```

```python
print(df.isnull().sum())
print(df.info())
```

# Descriptive analysis

```python
print("\nDescriptive Statistics")
print(df.describe())
```

# Data Visualization
## 1) Histograms


```python
# univariate analysis
plt.figure(figsize=(10, 6))
df.hist(bins=16, edgecolor='green', linewidth=1.9)
plt.suptitle('Histograms', size=20)
plt.show()
```

## 2) Box Plots

```python
plt.figure(figsize=(15, 7))
plt.subplot(2,3,1)
sns.boxplot(data=df['Age'],orient='h',palette="viridis")
plt.title('Box Plot of Age', size=16)
plt.xlabel('Age in Years')
plt.ylabel('')
plt.subplot(2,3,2)
sns.boxplot(data=df['Duration of Symptoms (months)'],orient='h',palette="plasma")
plt.title('Box Plot of Duration of symptoms', size=16)
plt.xlabel('Duration in months')
plt.ylabel('')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Y-BOCS Score (Obsessions)','Y-BOCS Score (Compulsions)']],orient='h',palette="plasma")
plt.title('Box Plot of Duration of symptoms', size=16)
plt.xlabel('Y-BOCS score')
plt.ylabel('')
plt.show()
```

## 3) Scatter plots 

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid",palette="bright")
sns.scatterplot(x='Y-BOCS Score (Obsessions)',y='Y-BOCS Score (Compulsions)',data=df, hue= 'Compulsion Type')
plt.title('Obsessions vs compulsion',size=16)
plt.show()
```

# Correlation Analysis

```python
num_columns =df.select_dtypes(include=['float','int'])
corr_matrix= num_columns.corr()
print('Correlation Matrix:')
print(corr_matrix)

df_1 = df.drop(['Age'],axis=1)
no_col =df_1.select_dtypes(include=['int','float'])
correlation_matrix= no_col.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix,annot=True, linecolor='black', cmap='Set2', linewidths=0.1)
plt.title('Heat Map', size=30)
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot= True, linecolor='green', cmap='Set2',linewidth=0.2)
plt.title('Heat Map for Correlation Analysis', size= 25)
plt.show()
```

# Univariate Analysis

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Duration of Symptoms (months)'],bins=19,color='green', edgecolor='blue',linewidth=1.4,kde=True)
plt.title('Duration of Symptoms')
plt.show()
```

# Bivariate Analysis

#### Relationship between two features

```python
import seaborn as sns 
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid', palette='bright')
sns.scatterplot(x='Duration of Symptoms (months)',y='Y-BOCS Score (Compulsions)',data= df,hue ='Family History of OCD', alpha=1)
plt.title('Duration of symptoms vs Y-BOCS score (compulsions)',size=16)
plt.show()

import seaborn as sns 
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid', palette='bright')
sns.scatterplot(x='Duration of Symptoms (months)',y='Y-BOCS Score (Compulsions)',data= df,hue ='Obsession Type', alpha=1)
plt.title('Duration of symptoms vs Y-BOCS score (compulsions)',size=16)
plt.show()
```

#### Relationship between a numerical and a categorical feature

```python
plt.figure(figsize=(10,7))
sns.set(style='whitegrid',palette='pastel')
sns.boxplot(x='Compulsion Type',y='Age',data=df)
plt.title('Age vs Compulsion Type',size=16)
plt.show()
print("\n")

plt.figure(figsize=(10,6))
sns.boxplot(x='Obsession Type',y='Age',data=df,hue='Gender', palette='Set1')
plt.title('Age vs Obsession Type',size=16)
plt.show()
```
